0) Необходимые библиотеки для решения поставленых задач

```python
import sympy
import numpy as np
from sympy.plotting import plot
from sympy import symbols
```

1) Определить корни

```python
x = symbols('x')

sympy.solveset(sympy.Eq(-12*x**4*sympy.sin(sympy.cos(x)) - 18*x**3 + 5*x**2 + 10*x - 30, 0), x)
```

2) Найти интервалы, на которых функция возрастает

Интервалов возрастания бесконечное множество.

3) Найти интервалы, на которых функция убывает

Интервалов убывания бесконечное множество.

4) Построить график

```python
plot(-12*x**4*sympy.sin(sympy.cos(x)) - 18*x**3 + 5*x**2 + 10*x - 30, (x, -100, 100))
```

5) Вычислить вершину

Присутствует бесконечное множество вершин

6) Определить промежутки, на котором f > 0 (взят диапазон -50 < x < 50 c шагом 1)

```python
x = np.arange(-50, 50, 1)
func = -12*x**4 * (np.sin(np.cos(x))) - 18*x**3 + 5*x**2 + 10*x - 30

vozrast_list = []
ubyvaet_list = []

for i in x:
    res = round(-12*i**4 * (np.sin(np.cos(i))) - 18*i**3 + 5*i**2 + 10*i - 30, 3)
    if res > 0: vozrast_list.append(res)
    else: ubyvaet_list.append(res)

print(f'f > 0 в интервале: \n{vozrast_list[-1]} --> {vozrast_list[0]}\n')
print(vozrast_list)
```

7) Определить промежутки, на котором f > 0 (взят диапазон -50 < x < 50 c шагом 1)

```python
print(f'\nf < 0 в интервале:\n{ubyvaet_list[0]} --> {ubyvaet_list[-1]}\n')
print(ubyvaet_list)
```
